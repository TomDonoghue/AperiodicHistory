This notebooks explores counts of papers reporting particular terms over time.

### Initialize LISC Object

```python
# Initialize counts object
ap_years = Counts1D()
```

```python
# Add terms to object
ap_years.add_terms(ALL_TERMS)
```

```python
# Check terms
ap_years.check_terms()
```

### Literature Collection

```python
# Define range of years to collect data for
start_year, end_year = 1945, 2015
years = list(range(start_year, end_year + 1))
```

```python
# Collect count of the number of aperiodic related papers per year
if RUN_COLLECTION:
    ap_years_results = collect_across_time(ap_years, years)
```

### Results

```python
# Plot the results over time of the results
plot_results_across_years(ap_years_results, 
                          xlim=[1945, 2015], 
                          ylim=[0, 12],
                          figsize=(11, 4))
```
